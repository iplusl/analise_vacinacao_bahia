# 04 - Análisando Hipóteses

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

## 04.01 O PNI está vacinando em maior ou menor quantidade nos últimos 4 anos?

Enquanto São Paulo foi o estado que mais aplicou doses no Brasil, quando fazemos o ranking de Doses per capita ele não chegou nem na metade. No ranking de doses per capita os três primeiros foram Maranhão, Alagoas e Amazonas, respectivamente.

Para analisarmos essa hipotése podemos olhar tanto para os dados de Cobertura Vacinal quanto para os dados de Doses aplicadas. Vamos analisar primeiramente através da Cobertura Vacinal.

```python
cv_por_uf_ano = pd.read_csv("../dados/dados limpos/cv_por_uf_ano.csv")
cv_por_uf_ano = cv_por_uf_ano.set_index("Unidade da Federação")
cv_por_uf_ano.head()
```

```python
cv_por_uf_ano.mean().mean()
```

Vamos plotar um gráfico para todo o país com a Média de Cobertura Vacinal ao longo do tempo, também vamos adicionar uma linha vermlha representando a média de todos os anos.

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=cv_por_uf_ano.mean())
ax.axhline(cv_por_uf_ano.mean().mean(), color="red")
plt.title("Média Cobertura Vacinal Brasil x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Cobertura Vacinal", fontdict = {"fontsize":12})
plt.ylim(0, 100)
plt.grid(linestyle='--')
plt.show()
```

Podemos ver no gráfico que nos últimos 4 anos a média da Cobertura Vacinal caiu bastante, desde 2016 estamos com uma média de cobertura abaixo da linha vermelha, ou seja, abaixo da média de todos os anos. Vamos ver se isso se repete com os dados da quantidade de doses.

```python
doses_por_uf_ano = pd.read_csv("../dados/dados limpos/doses_por_uf_ano.csv")
doses_por_uf_ano = doses_por_uf_ano.set_index("Unidade da Federação")
doses_por_uf_ano.head()
```

Vamos plotar um gráfico parecido com o anterior, com a soma de todo o país ao longo do tempo, também vamos adicionar uma linha vermlha representando a média de todos os anos.

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=doses_por_uf_ano.sum())
ax.axhline(doses_por_uf_ano.sum().mean(), color="red")
ticks_loc = ax.get_yticks().tolist()
ax.yaxis.set_major_locator(ticker.FixedLocator(ticks_loc))
ax.set_yticklabels(['{:,.0f}'.format(x) + ' Milhões' for x in ax.get_yticks()/1000000])
plt.ylim(0, 85000000)
plt.title("Doses Aplicadas no Brasil x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Doses Aplicadas", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

O mesmo comportamento que vimos com a cobertura vacinal também se repete com a quantidade de doses aplicadas. Os últimos 4 anos do nosso gráfico está abaixo da média de todos os anos.

## 04.02 É possível notar um decrescimento nos casos da vacina mais aplicada no Brasil pelo PNI?

Primeiramente vamos ver qual a vacina mais aplicada no Brasil e ver qual a doença que ela combate.

```python
imuno_cv_doses = pd.read_csv("../dados/dados limpos/imuno_cv_doses.csv")
imuno_cv_doses = imuno_cv_doses.set_index("Imuno")
imuno_cv_doses.head()
```

Vamos analisar para a quantidade de doses aplicadas.

```python
imuno_cv_doses = imuno_cv_doses.sort_values(by="Doses_Cáculos_CV", ascending=False)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=imuno_cv_doses, x=imuno_cv_doses.index, y="Doses_Cáculos_CV", palette=sns.color_palette("rocket_r", 23))
ticks_loc = ax.get_yticks().tolist()
ax.yaxis.set_major_locator(ticker.FixedLocator(ticks_loc))
ax.set_yticklabels(['{:,.0f}'.format(x) + ' Milhões' for x in ax.get_yticks()/1000000])
plt.title("Doses Aplicadas x Imuno - Brasil", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Imuno", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Doses Aplicadas", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

Com pouco mais de 80 milhões de doses aplicadas, a vacina BCG lidera o ranking. De acordo com o site da [Eurofarma](https://eurofarma.com.br/artigos/para-que-serve-a-vacina-bcg) a vacina BCG serve para combater doenças como a meningite tuberculosa e a tuberculose miliar. Vamos importar e analisar os casos de internações por tuberculose.

```python
tuberculose = pd.read_csv("../dados/dados limpos/tuberculose.csv")
tuberculose = tuberculose.set_index("Unidade da Federação")
tuberculose.head()
```

```python
tuberculose.sum()
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=tuberculose.sum())
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=7))
plt.ylim(0, 1300)
plt.title("Internações Tuberculose no Brasil x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Internações Tuberculose", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

Análisando o gráfico não possível perceber nenhum padrão, seja uma tendência, sazonalidade. Caso tivesse algum padrão poderiamos ter tentado relacionar com a vacinação do BCG. Muito provavelmente, por falta de conhecimento da área, escolhi um procedimento errado para esse dataset e com isso pode ser que esses dados não tenham nada a ver com a tuberculose e a meningite combatida pelo BCG.

## 04.03 As 3 vacinas mais aplicadas na Bahia também são as 3 mais aplicadas no Brasil?

Para ver quais as 3 vacinas mais aplicadas no Brasil podemos repetir o gráfico que fizemos agora há pouco.

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=imuno_cv_doses, x=imuno_cv_doses.index, y="Doses_Cáculos_CV", palette=sns.color_palette("rocket_r", 23))
ticks_loc = ax.get_yticks().tolist()
ax.yaxis.set_major_locator(ticker.FixedLocator(ticks_loc))
ax.set_yticklabels(['{:,.0f}'.format(x) + ' Milhões' for x in ax.get_yticks()/1000000])
plt.title("Doses Aplicadas x Imuno", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Imuno", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Doses Aplicadas", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

Através do gráfico podemos ver que as 3 vacinas mais aplicadas no Brasil são: BCG, Poliomelite e DTP. A vacina da BCG, como vimos antes, combate doenças como a meningite tuberculosa e a tuberculose miliar. A vacina da Poliomelite, como o nome já diz, combate a poliomelite, mais conhecida como paralisia infantil. Já a vacina DTP combate doenças como difteria, tétano e coqueluche.

Vamos analisar as vacinas da Bahia.

```python
imuno_cv_doses_ba = pd.read_csv("../dados/dados limpos/imuno_cv_doses_ba.csv")
imuno_cv_doses_ba = imuno_cv_doses_ba.set_index("Imuno")
imuno_cv_doses_ba.head()
```

Vamos analisar para a quantidade de doses aplicadas.

```python
imuno_cv_doses_ba = imuno_cv_doses_ba.sort_values(by="Doses_Cáculos_CV", ascending=False)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(data=imuno_cv_doses_ba, x=imuno_cv_doses_ba.index, y="Doses_Cáculos_CV", palette=sns.color_palette("rocket_r", 23))
ticks_loc = ax.get_yticks().tolist()
ax.yaxis.set_major_locator(ticker.FixedLocator(ticks_loc))
ax.set_yticklabels(['{:,.0f}'.format(x) + ' Milhões' for x in ax.get_yticks()/1000000])
plt.title("Doses Aplicadas x Imuno - Bahia", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Imuno", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Doses Aplicadas", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

Comparando a Bahia com o Brasil, vemos que o top 3 vacinas aplicadas se mantem no mesmo padrão.

## 04.04 É possível notar um decrescimento nos casos de poliomelite na Bahia?

```python
polio = pd.read_csv("../dados/dados limpos/polio.csv")
polio = polio.set_index("Unidade da Federação")
polio.head()
```

```python
polio.loc["29 Bahia"]
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=polio.loc["29 Bahia"])
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=3))
plt.ylim(0, 3)
plt.title("Internações Poliomelite Aguda na Bahia x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Internações Poliomelite Aguda", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

Análisando o gráfico não possível perceber nenhum padrão já que há poquíssimos casos de internações por poliomelite aguda. Talvez o ideal ao analisar esse tipo de doença seja olhar para os dados de produção ambulatorial, onde pode ser que os casos que ainda existem de poliomelite não conseguem se agravar ao ponto de exigir uma internação.

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=polio.sum())
ax.xaxis.set_major_locator(ticker.IndexLocator(base=12, offset=3))
plt.ylim(0, 10)
plt.title("Internações Poliomelite Aguda no Brasil x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Internações Poliomelite Aguda", fontdict = {"fontsize":12})
plt.grid(linestyle='--')
plt.show()
```

O mesmo comportamento se repete para a esfera nacional, onde ao longo dos anos tivemos poucos casos de internações de poliomelite aguda. Somente em janeiro de 2008 que no Brasil houve um pico chegando a 9 internações.
