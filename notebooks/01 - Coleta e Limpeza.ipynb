# 01 - Coleta e Limpeza

Os primeiros dados coletados serão as Coberturas Vacinais por Unidade da Federação e Ano.

## Coberturas Vacinais por Unidade da Federação e Ano

Nesse dataset teremos o percentual da Corbetura Vainal, onde em cada linha teremos uma Unidade da Federação e em cada coluna o ano.

```python
import pandas as pd
```

```python
cv_por_uf_ano = pd.read_csv("../dados/A100607189_28_143_208.csv",
            encoding="ISO-8859-1",
            skiprows=3, sep=";", skipfooter=20,
            thousands=".", decimal=",", engine="python")
cv_por_uf_ano
```

Agora que já temos os dados inseridos vamos eliminar os valores ausentes.

```python
cv_por_uf_ano.isna().sum()
```

Temos valores ausentes apenas nos primeiros anos (1994, 1995 e 1996), vamos eliminar essas colunas.

```python
cv_por_uf_ano = cv_por_uf_ano.dropna(axis="columns")
cv_por_uf_ano.head()
```

```python
cv_por_uf_ano.isna().sum().sum()
```

Além disso também vamos remover a coluna Total.

```python
cv_por_uf_ano = cv_por_uf_ano.drop("Total", axis="columns")
cv_por_uf_ano.head()
```

Agora vamos salvar o arquivo para utilizarmos mais tarde.

```python
cv_por_uf_ano.to_csv("../dados/dados limpos/cv_por_uf_ano.csv", index=False)
```

## População

Vamos pegar os dados de população por estado disponíveis no [Wikipédia](https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o)

```python
populacao = pd.read_html('https://pt.wikipedia.org/wiki/Lista_de_unidades_federativas_do_Brasil_por_popula%C3%A7%C3%A3o')
populacao = populacao[0]
populacao = populacao.drop(["Posição", "% da pop. total", "País comparável (habitantes)"], axis="columns")
populacao = populacao.rename({"Unidade federativa":"Unidade da Federação"}, axis="columns")
populacao = populacao.set_index("Unidade da Federação")
populacao["População"] = populacao["População"].str.replace(" ", "").astype("int")
populacao.head()
```

```python
populacao.to_csv("../dados/dados limpos/populacao.csv")
```

## Casos Tuberculose

Nesse dataset teremos em cada linha o estado, e cada coluna representa um mês e ano.

```python
tuberculose = pd.read_csv("../dados/A234202189_28_143_208.csv",
            encoding="ISO-8859-1",
            skiprows=4, sep=";", skipfooter=12,
            thousands=".", decimal=",", engine="python")
tuberculose = tuberculose.set_index("Unidade da Federação")
tuberculose
```

Vamos remover a coluna Total já que ela pode atraplhar no gráficos.

```python
tuberculose = tuberculose.drop("Total", axis="columns")
```

Nesse caso, de acordo com a documentação do TABNET, os dados com valores "-" devem ser considerados como 0. Vamos fazer essa substituição.

```python
tuberculose = tuberculose.replace("-", 0)
tuberculose.head()
```

```python
tuberculose.sum()
```

Vamos verificar se tem algum dado faltante.

```python
tuberculose.isna().sum().sum()
```

Nesse caso não temos nenhum dado faltante. Vamos colocar as colunas como date para facilitar nossa analise mais tarde.

```python
from datetime import date

meses = {
    "Jan" : 1,
    "Fev" : 2,
    "Mar" : 3,
    "Abr" : 4,
    "Mai" : 5,
    "Jun" : 6,
    "Jul" : 7,
    "Ago" : 8,
    "Set" : 9,
    "Out" : 10,
    "Nov" : 11,
    "Dez" : 12
}

def para_dia (ano_mes: str):
    ano: int = int(ano_mes[:4])
    mes = meses[ano_mes[5:]] 
    return date(ano, mes, 1)

tuberculose.columns = tuberculose.columns.map(para_dia)
tuberculose.head()
```

Agora vamos salvar o arquivo para analisarmos posteriormente.

```python
tuberculose.to_csv("../dados/dados limpos/tuberculose.csv", index=True)
```
