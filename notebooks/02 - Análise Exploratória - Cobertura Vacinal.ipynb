# 02 - Análise Exploratória - Cobertura Vacinal

## Coberturas Vacinais por Unidade da Federação e Ano

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cv_por_uf_ano = pd.read_csv("../dados/dados limpos/cv_por_uf_ano.csv")
cv_por_uf_ano = cv_por_uf_ano.set_index("Unidade da Federação")
cv_por_uf_ano.head()
```

A Cobertura Vacinal é percentual de menores de um ano de idade imunizados com vacinas específicas, em determinado local e período. Vamos ver como a média da cobertura vacinal se comporta ao longo dos anos.

```python
cv_por_uf_ano.mean()
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=cv_por_uf_ano.mean())
plt.title("Média Cobertura Vacinal Brasil x Tempo", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Ano", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Cobertura Vacinal", fontdict = {"fontsize":12})
plt.ylim(0, 100)
plt.grid(linestyle='--')
plt.show()
```

Podemos observar que a média de cobertura vacinal do Brasil se manteve estável perto dos 80% maior parte do tempo. Em 2015 ela teve seu maior pico atingindo 93.65% e logo em seguida uma grande queda em 2016 atingindo 52.72% de cobertura vacinal.

De acordo o [site da Fundação Oswaldo Cruz](https://portal.fiocruz.br/noticia/revista-radis-aborda-queda-da-cobertura-vacinal-no-brasil#:~:text=Mas%2C%20tanto%20em%202016%20quanto,%2C4%25%2C%20em%202017.), tanto em 2016 quanto em 2017 a cobertura da vacina contra a pólio ficou mais de 10 pontos percentuais abaixo da meta, além disso também houve uma queda nas principais vacinas ofertadas gratuitamente à população.
<br> Podemos ver claramente essa queda no gráfico.

```python
cv_por_uf_ano.T.mean().sort_values(ascending = False)
```

```python
plt.figure(figsize=(10,6))
ax = sns.lineplot(data=cv_por_uf_ano.T.mean())
plt.title("Média Cobertura Vacinal x Estado", fontdict = {"fontsize":16, "fontweight":"bold"})
plt.xlabel("Estado", fontdict = {"fontsize":12})
plt.xticks(rotation=45, ha="right")
plt.ylabel("Cobertura Vacinal", fontdict = {"fontsize":12})
plt.ylim(0, 100)
plt.grid(linestyle='--')
plt.show()
```

Podemos observar que a média de cobertura vacinal dos estados se manteve estável entre 70% e 80% durante maior parte do tempo.
<br>Os três estados com maiores médias são, respectivamente, Rondônia, Mato Grosso do Sul e Goiás. A Bahia ocupa a 19ª posição nesse ranking.
